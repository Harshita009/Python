```python
#importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the Dataset
df = pd.read_csv("D:/Documents/Documents/Besant technology/5.python/ppt/Vyankat sir notes/titanic.csv")
```

```python
#it will show first 5 rows of datasets
df.head()
```

```python
#Get Shape of the Dataset Rows and Columns
rows, cols = df.shape
print(f"Number of rows: {rows}")
print(f"Number of columns: {cols}")
```

```python
#here we can get column names and data types
print("Column Names and Data Types:")
print(df.dtypes)
```

```python
# here it show average age of passengers
average_age = df['age'].mean()
print(f"Average age of the passengers: {average_age:.2f}")
```

```python
#Number of Passengers Who Survived  it will show
survived_count = df['survived'].sum()
print(f"Number of passengers who survived: {survived_count}")
```

```python
#Number of Males and Females
gender_counts = df['sex'].value_counts()
print("Gender counts:")
print(gender_counts)
```

```python
common_embark = df['embarked'].mode()[0]
print(f"Most common embarkation point: {common_embark}")
```

```python
gender_counts = df['sex'].value_counts()

plt.bar(gender_counts.index, gender_counts.values, color=['green', 'orange'])
plt.title("Number of Male vs Female Passengers")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
```

```python
plt.hist(df['age'].dropna(), bins=20, edgecolor='black')
plt.title("Distribution of Passenger Ages")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
survival_counts = df['survived'].value_counts()
labels = ['not survived', 'survived']
plt.pie(survival_counts, labels=labels, autopct='%1.1f%%', startangle=90, colors=['yellow', 'purple'])
plt.title("survival Distribution")
plt.axis('equal')
plt.show()
```

```python
class_counts = df['pclass'].value_counts()

plt.bar(class_counts.index.astype(str), class_counts.values, color='teal')
plt.title("Number of Passengers in Each Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.show()

```
